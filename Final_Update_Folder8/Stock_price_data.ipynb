Next, we divide these three dataframes into training and testing sets. For each of them, we divide it into 80% training set and 20% testing set. For each set (training or testing), we set 50 time points. Based on these 50 time points, we continue to divide the set into different batches. Then we will have the consistent format three-dimension tensor (batch number, time points number, stock number).

____________________________________________________________________________________________________

[For Dow 30]:

```python
df_dow_train = df_dow[:round(len(df_dow)*0.8)].reset_index()
df_dow_train = df_dow_train.drop('index',axis = 1)
df_dow_train
```

```python
df_dow_test = df_dow[round(len(df_dow)*0.8):].reset_index()
df_dow_test = df_dow_test.drop('index',axis = 1)
df_dow_test
```

```python
# We adjust the dimensions of both the training and testing sets, from 1-D to 3-D:
# We split the original training and testing sets into equal-length fragments. By this way, each fragment has 50 or 51 time points and we just take the first 50 time points as the time series. 
# For the training set , there are 2014 rows so we split it into 40 fragments then each fragment has 50 or 51 time points.
# For the testing set , there are 503 rows so we split it into 10 fragments then each fragment has 50 or 51 time points.

dow_data_train = []
for i in range(len(np.array_split(df_dow_train.iloc[:,1:],40))):  
  batch_group = []
  for ii in range(len(np.array_split(df_dow_train.iloc[:,1:],40)[i])):
    batch_group.append(list(np.array_split(df_dow_train.iloc[:,1:],40)[i].iloc[ii]))
  batch_group = np.array(batch_group[:50])  # We just take the first 50 rows as 50 time points. By this way, we can ensure all arrays have the same dimensions: (50, 29) in one batch
  dow_data_train.append(np.array(batch_group.reshape((50,len(np.array_split(df_dow_train.iloc[:,1:],40)[i].columns)))))

dow_data_test = []
for i in range(len(np.array_split(df_dow_test.iloc[:,1:],10))):  
  batch_group = []
  for ii in range(len(np.array_split(df_dow_test.iloc[:,1:],10)[i])):
    batch_group.append(list(np.array_split(df_dow_test.iloc[:,1:],10)[i].iloc[ii]))
  batch_group = np.array(batch_group[:50])  # We just take the first 50 rows as 50 time points. By this way, we can ensure all arrays have the same dimensions: (50, 29) in one batch
  dow_data_test.append(np.array(batch_group.reshape((50,len(np.array_split(df_dow_test.iloc[:,1:],10)[i].columns)))))
```

```python
# The the 3 dimensions are correct:
# 40 or 10 batches, 50 time points, 29 features (stocks)

print(np.array(dow_data_train).shape)
print(np.array(dow_data_test).shape)
```
